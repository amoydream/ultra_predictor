```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegresso
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
df = pd.read_csv('./prediction_csv/prediction_all.csv')
```

```python
df
```

```python
df
```

```python
 df.dropna(subset=['best_ten_run_in_hours'], inplace=True)
```

```python
df
```

```python
df = df[(np.abs(stats.zscore(df['best_ten_run_in_hours'])) < 3)]
```

```python
df = df[(np.abs(stats.zscore(df['time_result_in_hours'])) < 3)]
```

```python
df
```

```python
dummy_sex = pd.get_dummies(df['runner_sex'])
```

```python
df = pd.concat([df, dummy_sex], axis=1)
```

```python
df
```

```python
dummy_month = pd.get_dummies(df['month_of_the_race'])
```

```python
df = pd.concat([df, dummy_month], axis=1)
```

```python
df
```

```python
months = {"January","February","March","April","May","June","July","August","September","October","November","December"}
```

```python
 months_to_add = list(months.intersection(set(df.columns)))
```

```python
X = df[months_to_add + ['M', 'W', 'best_ten_run_in_hours', 'runner_age','itra_point','food_point','time_limit','distance','elevation_gain','elevation_lost']]
```

```python
X
```

```python
y = df[['time_result_in_hours']]
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.2, random_state=201)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
lm
```

```python

print(lm.intercept_)
```

```python
lm.coef_[0]
```

```python
cdf = pd.DataFrame(lm.coef_[0], X.columns, columns=['Coeff'])
```

```python
cdf
```

```python
prediction = lm.predict(X_test)
```

```python
prediction
```

```python
X_test
```

```python
plt.scatter(y_test, prediction)
```

```python
sns.distplot((y_test-prediction), bins=35)
```

```python
np.sqrt(metrics.mean_squared_error(y_test, prediction))
```
