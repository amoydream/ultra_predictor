```python
from elasticsearch import Elasticsearch
from pandasticsearch import Select
import pandas as pd
```

```python
elastic_client =  Elasticsearch('http://elasticsearch:9200')
```

```python
data = elastic_client.search(index="race_results",scroll='2m', size=1000, 
                            body={
                                "query":{
                                            "exists": { "field": "best_ten_run_in_hours"}
                                        }
                                 
                                 }
                            )
```

```python
sid = data['_scroll_id']

```

```python
scroll_size = len(data['hits']['hits'])
```

```python
frames = []
```

```python
while scroll_size > 0:
    print('scroolling')
    pandas_df = Select.from_dict(data).to_pandas()
    frames.append(pandas_df)
    data = elastic_client.scroll(scroll_id=sid, scroll='2m')
    sid = data['_scroll_id']
    scroll_size = len(data['hits']['hits'])
    print(scroll_size)
```

```python
result = pd.concat(frames)
```
